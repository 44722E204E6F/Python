### **Algoritmo**
Calcula área y perímetro de figuras bidimencioneles y el volumen de poliedros regulares.
<br><br><br>
**NOTA**: *Debido a la gran cantidad de figuras bidimencioneles existentes y su complejidad, solo se realizan los calculos de alguna de las más conocidas.*

**Librerias**

```python
### Librerias
import math
```

**Funciones**

```python
### Validación de entradas
def get_pos_float(prompt):
    while True:
        try:
            v = float(input(prompt))
            if v > 0: return v
            print("\n\t❌ Ingrese un número positivo.")
        except:
            print("\n\t❌ Entrada inválida.")

def get_int(prompt, minv=None, maxv=None):
    while True:
        try:
            v = int(input(prompt))
            if (minv is None or v >= minv) and (maxv is None or v <= maxv):
                return v
            print(f"\n\t⚠️ Ingrese entero entre {minv} y {maxv}.")
        except:
            print("\n\t⚠️ Entrada inválida.")

### Funciones de área y perímetro
def cuadrado(a): return a*a, 4*a

def rectangulo(w,h): return w*h, 2*(w+h)

def circulo(r): return math.pi*r*r, 2*math.pi*r

def triangulo_bh(b,h): return 0.5*b*h, None

def triangulo_Heron(a,b,c):
    s=(a+b+c)/2; A=s*(s-a)*(s-b)*(s-c)
    return (math.sqrt(A) if A>0 else 0.0), a+b+c

def paralelogramo(b,h): return b*h, 2*(b+h)  # Romboide

def rombo(d1,d2):
    area = d1*d2/2
    per = math.sqrt((d1/2)**2 + (d2/2)**2)
    return area, 4*per

def delta(d1,d2,l1,l2):
    return d1*d2/2, (2*l1)+(2*l2)

def trapezoide(b1,b2,h,legs_known=False,leg1=None,leg2=None):
    area = 0.5*(b1+b2)*h
    if legs_known:
        per = b1+b2+leg1+leg2
    else:
        # asume isósceles para calcular lados
        leg = math.sqrt(((b1-b2)/2)**2 + h*h)
        per = b1+b2+2*leg
    return area, per

def poligono_regular(n,a): return (n*a*a)/(4*math.tan(math.pi/n)), n*a

def elipse(a,b):
    area = math.pi*a*b
    per = math.pi*(3*(a+b)-math.sqrt((3*a+b)*(a+3*b)))  # Aproximación Ramanujan
    return area, per

# Circulares
def sector(R,theta): return 0.5*R*R*theta, R*theta + 2*R

def segmento_circular(R,h):
    if h<=0 or h>2*R: return 0.0, None
    theta = 2*math.acos((R-h)/R)
    area = 0.5*R*R*(theta - math.sin(theta))
    chord = 2*R*math.sin(theta/2)
    per = chord + R*theta
    return area, per

def anillo(R,r): return math.pi*(R*R - r*r), 2*math.pi*(R + r)

def sector_anular(R,r,theta):
    area = 0.5*(R*R - r*r)*theta
    per = R*theta + r*theta + 2*(R - r)
    return area, per


### Volúmenes poliedros regulares
def vol_tetra(a): return a**3/(6*math.sqrt(2))
def vol_cube(a): return a**3
def vol_octa(a): return (math.sqrt(2)/3)*a**3
def vol_dode(a): return ((15+7*math.sqrt(5))/4)*a**3
def vol_icos(a): return (5*(3+math.sqrt(5))/12)*a**3
```

**Menús**

```python

### Menú figuras bidimencionales
def menu_figuras():
    items = [
      "Cuadrado","Rectángulo","Círculo","Triángulo (base/altura)","Triángulo (Herón)",
      "Romboide (paralelogramo)","Rombo (diagonales)","Deltoide (diagonales)","Trapecio",
      "Polígono regular","Elipse","Sector circular","Segmento circular","Corona (anillo)","Trapecio circular (sector anular)"
    ]
    while True:
        print("\n Figuras bidimencionales\n\t • Selecciona la opción deseada [0 volver]...\n")
        for i,name in enumerate(items,1): print(f"\t{i}. {name}")
        opt = get_int("\n\t  » Opción: ",0,len(items))
        if opt==0: break
        if opt==1:      # Cuadrado
            a=get_pos_float("\n\t ◦ Lado: "); A,P = cuadrado(a)
        elif opt==2:    # Rectangulo
            w=get_pos_float("\n\t ◦ Ancho: "); h=get_pos_float("\t ◦ Alto: "); A,P = rectangulo(w,h)
        elif opt==3:    # Circulo
            r=get_pos_float("\n\t ◦ Radio: "); A,P = circulo(r)
        elif opt==4:    # Triangulo (base-altura)
            b=get_pos_float("\n\t ◦ Base: "); h=get_pos_float("\t ◦ Altura: "); A,P = triangulo_bh(b,h)
        elif opt==5:    # Triangulo de Heron
            a=get_pos_float("\n\t ◦ Lado a: "); b=get_pos_float("\t ◦ Lado b: "); c=get_pos_float("\t ◦ Lado c: "); A,P = triangulo_Heron(a,b,c)
        elif opt==6:    # Paralelogramo
            b=get_pos_float("\n\t ◦ Base: "); h=get_pos_float("\t ◦ Altura: "); A,P = paralelogramo(b,h)
        elif opt==7:    # Rombo
            d1=get_pos_float("\n\t ◦ Diagonal mayor: "); d2=get_pos_float("\t ◦ Diagonal menor: "); A,P = rombo(d1,d2)
        elif opt==8:    # Deltoide
            d1=get_pos_float(prompt="\n\t ◦ Diagonal principal: "); d2=get_pos_float(prompt="\t ◦ Diagonal secundaria: ")
            l1=get_pos_float(prompt="\t ◦ Lado mayor: "); l2=get_pos_float(prompt="\t ◦ Lado menor: "); A,P = delta(d1,d2,l1,l2)
        elif opt==9:    # Trapezoide
            b1=get_pos_float("\n\t ◦ Base mayor: "); b2=get_pos_float("\t ◦ Base menor: "); h=get_pos_float("\t ◦ Altura: ")
            iso = get_int("\n\t¿Asumir isósceles para perímetro?\n\t1. Sí\n\t0. No\n\t » Opción: ",0,1)
            if iso:
                A,P = trapezoide(b1,b2,h,False)
            else:
                l1=get_pos_float("\n\t ◦ Lado 1: "); l2=get_pos_float("\t ◦ Lado 2: "); A,P = trapezoide(b1,b2,h,True,l1,l2)
        elif opt==10:   # Poligono regular
            n=get_int("\n\t ◦ Número de lados (>=3): ",3); a=get_pos_float("\t ◦ Arista: "); A,P = poligono_regular(n,a)
        elif opt==11:   # Elipso
            a=get_pos_float("\n\t ◦ Semi-eje mayor a: "); b=get_pos_float("\t ◦ Semi-eje menor b: "); A,P = elipse(a,b)
        elif opt==12:   # Sector circular
            R=get_pos_float("\n\t ◦ Radio: "); unit = get_int("\t ◦ Ángulo en:\n\t\t1. Grados\n\t\t2. Radianes:\n\t\t » ",1,2)
            theta = math.radians(get_pos_float("\t ◦ Ángulo grados: ")) if unit==1 else get_pos_float("\t ◦ Ángulo radianes: ")
            A,P = sector(R,theta)
        elif opt==13:   # Segmento circular
            R=get_pos_float("\n\t ◦ Radio: "); h=get_pos_float("\t ◦ Altura segmento [h]: "); A,P = segmento_circular(R,h)
        elif opt==14:   # Anillo
            R=get_pos_float("\n\t ◦ Radio mayor: "); r=get_pos_float("\t ◦ Radio menor: "); A,P = anillo(R,r)
        elif opt==15:   # Sector anular
            R=get_pos_float("\n\t ◦ Radio mayor: "); r=get_pos_float("\t ◦ Radio menor: ")
            unit = get_int("\n\tÁngulo en:\n\t\t1. Grados\n\t\t2. Radianes\n\t\t » Opción: ",1,2)
            theta = math.radians(get_pos_float("\t\t ◦ Ángulo grados: ")) if unit==1 else get_pos_float("\t\t ◦ Ángulo radianes: ")
            A,P = sector_anular(R,r,theta)
        print(f"\n ■ Área: {A:.6f}", f"\n □ Perímetro: {P:.6f}" if P else "\n □ Perímetro: (requiere lados)")

### Menú poliedros regulares
def menu_solidos():
    names = ["Tetraedro","Cubo","Octaedro","Dodecaedro","Icosaedro"]
    funcs = [vol_tetra,vol_cube,vol_octa,vol_dode,vol_icos]
    while True:
        print("\n Poliedros regulares\n\t • Selecciona la opción deseada [0 volver]...\n")
        for i,n in enumerate(names,1): print(f"\t{i}. {n}")
        opt = get_int("\n\t » Opción: ",0,len(names))
        if opt==0: break
        a=get_pos_float("\n Arista a: "); V = funcs[opt-1](a)
        print(f" Volumen {names[opt-1]}: {V:.6f}")
```

**Main**

```python

### Menú principal
def main():
    print("Calculadora: áreas, perímetros y volúmenes (poliedros regulares)")
    while True:
        print("\n\t1. Figuras bidimencionales\n\t2. Poliedros regulares\n\t0. Salir")
        o = get_int("\n\t » Opción: ",0,2)
        if o==0: break
        if o==1: menu_figuras()
        else: menu_solidos()
    print("\n\n\t☁︎⏾ᶻᶻᶻ")

if __name__=="__main__":
    main()
```

###**Notas importantes**

  **Ángulos**: el menú permite ingresar grados o radianes; internamente se usan radianes.

  **Perímetros**: cuando faltan lados (p. ej. deltoide por diagonales) el programa devuelve Perímetro: (requiere lados) y solicita lados si quieres calcularlo.

  **Unidades**: usa unidades coherentes; áreas en unidades², volúmenes en unidades³.
